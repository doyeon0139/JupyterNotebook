```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

path = "#zg-ordered-list > li"

sell = soup.select(path)

sell
```

```python
for se in sell:
    rank = se.select("li > span > div > div > span.a-size-small.aok-float-left.zg-badge-body.zg-badge-color > span")[0].text
    tit = se.select(" li > span > div > span > a > div")[0].text
    re = se.select("li > span > div > span > div > a.a-size-small.a-link-normal")[0].text
    price = se.select( "a.a-link-normal.a-text-normal")[0].text
   
        
    
    print("순위 : ", rank)
    print("타이틀 : ", tit)
    print("댓글수 : ", re)
    print("가격 : ", price)
  
    print("="*100)
        

        
```
