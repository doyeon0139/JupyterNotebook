```python
import requests
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import warnings
warnings.filterwarnings(action='ignore')
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://cafe.naver.com/joonggonara")
```

```python
elem = driver.find_element_by_class_name("inp")

s = input("검색어")

elem.send_keys(s)

driver.find_element_by_css_selector("#cafe-search > form > button").click()
# elem.send_keys(Keys.RETURN)
```

```python
html = driver.page_source

soup = BeautifulSoup(html, "html.parser")

path = "#main-area > div:nth-child(7) > table > tbody > tr"

cafe = soup.select(path)

cafe
```

```python
for kuku in cafe:
    tit = cucu.select("")[0].text
    per = cucu.select("")[0].text
    date = cucu.select("")[0].text
    count = cucu.select("")[0].text
    
    print(tit)
    print(per)
    print(date)
    print(count)
```
