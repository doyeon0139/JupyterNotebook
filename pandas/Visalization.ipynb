### Line Chart
* plot()
* 데이터 추세

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)    # plot 메소드
```

```python
x = [1,2,3,4,5]
y = [i**2 for i in x]

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.plot(x,y)
plt.plot(x1,y1)

plt.savefig('test_vis.jpg')        # 그래프 저장
```

```python
x1 = range(50)
y1 = [np.cos(value) for value in x1]

x2 = range(50)
y2 = [np.sin(value) for value in x2]

plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
plt.figure(figsize=(10,5))                                     # 도화지 만듬

plt.subplot(1,2,1)                                               #1행 2열로 도화지를 나눠서 1번째에
plt.plot(x1, y1, color = 'red', linestyle='dotted')             # 이것을 그림  색깔 라인스타일 지정

plt.subplot(1,2,2)                                               #1행 2열로 도화지를 나눠서 2번째에
plt.plot(x2, y2, color = 'blue', linestyle='dashed')             # 이것을 그림
```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value+50 for value in x2]

plt.plot(x1, y1, c ='green', ls = '-.', label = 'Line1')
plt.plot(x2, y2, c = 'black', ls = ':', label = 'Line2')

plt.text(50,60, 'Line1')            # 텍스트 입력
plt.annotate('Line 2', xytext = (30, 110), xy = (50,100), arrowprops=dict(facecolor='yellow', shrink=0.1)) # 화살표

plt.legend(fancybox=True, shadow=True, loc='upper right')    # 범례넣기 위치값 셋팅가능

plt.xlabel('X_axis')
plt.xlabel('Y_axis')    # 축이름 넣기?
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)   # 50개의 색깔을 랜덤으로 뽑아서 적용
area = np.pi * (30*np.random.rand(n))**2     # 데이터 크기에 따라? 크기가 달라짐?

plt.scatter(x,y, c=colors, s=area , alpha = 0.5) # alpha = 0.5 투명도
plt.savefig('text_img.jpg')
```

```python
plt.figure(figsize=(10,5))

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(url)
drinks.head()
```

```python
drinks.columns = ['country', 'beer','spirit','wine','litres','continent']
drinks.head()
```

```python
df = drinks.copy()
df
```

```python
df.plot()
```

```python
df.beer.plot(kind='bar')
```

```python
df.sort_values('beer')['beer'].plot(kind='bar')   # beer값을 오름차순으로 정렬, beer만 출력
```

```python
plt.figure(figsize=(15,3))
df.beer.sort_values(ascending=False)[:20].plot(kind='bar')
```
