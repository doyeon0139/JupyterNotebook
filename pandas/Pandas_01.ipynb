### missing value

```python
df = pd.DataFrame(np.random.randn(5,5), index=['a','b','c','d','e'], \
             columns=['one','two','three','four','five'])
```

```python
df.iloc[1:3, 1] = np.nan
df.iloc[2, 2:4] = np.nan
df
```

```python
# # nan, Nan, NaN 값 처리 방식

# 0. nan값을 탐지  => isnull(), isna()
# 1. nan값을 포함한 row 또는 column을 삭제  => dropna()
# 2. nan값을 다른 값으로 채우는 방법  => fillna()
```

```python
df.isnull()   # nan이면 True
```

```python
df.isnull().sum()   # df에 함수를 적용하면 기본 axis=0(세로방향)   ##sum(axix=0) 이 들어가 있음
```

```python
df.isnull().sum().sum()
```

```python
df.isnull().sum(axis=1)   # 가로방향
```

```python
# dropna()
```

```python
df.dropna()      # nan값이 잇는 row를 삭제
```

```python
df.dropna(axis=1)    # nan이 있는 column을 지우고 싶음
```

```python
df.dropna(how = 'all')   # 전체 데이터가 nan인 경우만 삭제
```

```python
df.dropna(thresh = 3)  # non을 가진 row의 데이터가 3개 이상이면 삭제 하지마
```

```python
#  fillna()
```

```python
df.fillna(0)          # non 값을 0으로 채우기
```

```python
df.fillna(method='ffill')       # 앞에있는 값으로 non값을 채우라
```

```python
df.fillna(method='bfill')       # 뒤에있는 값으로 non값을 채우라
```

```python
df.fillna(df.mean())          # column 평균값으로 채우기??????row??
```

```python
df.interpolate()   # ????? 평균채우기라는데 뭔 평균이라는지
```

### apply

```python
d = {'a':[10,20,30],'b':[20,30,40],'c':[50,60,70]}
dd = pd.DataFrame(d)
dd
```

```python
def tmpf(x):
    return x*2 +5
```

```python
dd.apply(tmpf)
```

```python
def sq(x):
    return x**2
```

```python
dd['a'].apply(sq)
```

```python
dd['a'].apply(lambda x: x**2)
```

### merge/ concat
* dataframe을 합치는 것
* merge는 sql의 table을 합하는 join과 같다

```python
from IPython
```

```python
# merge의 종류:
# 두 dataframe에 공통 칼럼이 존재하고

# A_df와 B_df를 병합하는 경우
# 1. inner merge : A_df 와 B_df 모두에 존재하는 데이터(row)들로만 합하는 방식
# 2. outer merge : A_df 와 B_df 어느 한 곳에 존대하는 데이터는 모두 포함해서 합하는 방식
# 3. left merge : A_df 의 데이터를 모두 포함, B_df에서 A_df에 존재하는 데이터만 포함
# 4. right merge : B_df 의 데이터를 모두 포함, A_df에서 B_df에 존재하는 데이터만 포함
```

```python
raw_data = {
    'subject_id': ['1','2','3','4','5','7','8','9','10','11'],
    'test_score': [51,13,16,71,63,74,90,88,87,99]
    
}
df_a = pd.DataFrame(raw_data)
df_a
```

```python
raw_data = {
    'subject_id' : ['4','5','6','7','8'],
    'first_name' : ['Billy','Brian','Bran','Bryce','Betty'],
    'last_name' : ['Bonder','Black','Balwner','Brice','Btisan']
    
}
df_b = pd.DataFrame(raw_data)
df_b
```

```python
df_a.head(2)
```

```python
df_b.head(2)
```

```python
df_a.subject_id.values
```

```python
df_b.subject_id.values
```

```python
pd.merge(df_a, df_b, on='subject_id', how = 'inner')   # 기준 서브젝트아이디 어떤것은 이너 
# pd.merge(df_a, df_b, on='subject_id')  디폴트가 이너라 안써도 이너
```

```python
pd.merge(df_a, df_b, on='subject_id', how='outer')  # outmerge : nan값이 생성되기 쉽다
```

```python
df_a.merge(df_b, on='subject_id', how='right')       
```

```python
df_a.merge(df_b, on='subject_id', how='left')        # 비어있는 내용은 nan표시
```

```python
df_a.merge(df_b, on='subject_id', how='right') 
```
