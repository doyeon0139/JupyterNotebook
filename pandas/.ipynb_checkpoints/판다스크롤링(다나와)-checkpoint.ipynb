```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
query = '무선청소기'
url = f'http://search.danawa.com/dsearch.php?query={query}&tab=main'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
type(soup)
```

```python
title_lst = []
price_lst = []
spec_lst = []

for lst in soup.select('#productListArea > div.main_prodlist.main_prodlist_list > ul > li.prod_item'): # li에 prod_item의 클래스를 가진 애들만 모으기(큰 의미는 없더라..)
    if lst.select('div.prod_main_info') and lst['class'] == ['prod_item']: # li 밑에 ad 어쩌구 제거 and li에 클래스가 prod_item으로 딱 끝나는것만 가져가기
        title = lst.select('p.prod_name > a')[0].text 
        price = lst.select('p.price_sect > a')[0].text
        spec = lst.select('div.spec_list')[0].text
        print(title)
        print(price)
        print(spec)
        title_lst.append(title)
        price_lst.append(price)
        spec_lst.append(spec)
        print('-'*60)
```

```python
# dataframe에 저장
df = pd.DataFrame({'title':title_lst,'price':price_lst,'spec':spec_lst})
```

```python
df
```

```python
# title에서 제조사 이름부분만 짤라내기
df['company'] = df['title'].apply(lambda x : x.split()[0])
```

```python
# title에서 제품 이름부분만 짤라내기(split때문에 분할된거 원상복구)
df['product'] = df['title'].apply(lambda x : x.split(' ', 1)[1])
```

```python
df
```

```python
# 뒤에 함수로 뺑뺑이 돌리려고 split으로 쪼개서 저장하기
df['spec_lst'] = df['spec'].apply(lambda x : x.split('/'))
```

```python
df
```

```python
# spec에서 사용시간이랑 흡입력 추출

def use(lst):
    for s in lst:
        if '사용시간' in s:
            return s.split(':')[1].strip(' ')
            
def suc(lst):
    for s in lst:            
        if '흡입력' in s:
            return s.split(':')[1].strip(' ')
        
df['usetime'] = df['spec_lst'].apply(use)
df['suction'] = df['spec_lst'].apply(suc)
```

```python
df
```

```python
# 가격 정수로 수정
df['price'] = df['price'].apply(lambda x : x.replace(',','').strip('원 ')).astype('int64')
df
```

```python
# 정규표현식 사용해서 시간값 분단위로 변경, 최대 시간을 분단위로 출력
import re
def convertime(time):
    if time:
        if '~' in time:
            time = time.split('~')[1]
        elif ',' in time:
            time = time.split(',')[1]
        hour = 0
        if '시간' in time:
            regex = re.compile('\d+시간')
            hour += int(re.search('\d+',regex.search(time).group()).group())*60
        if '분' in time:
            regex = re.compile('\d+분')
            hour += int(re.search('\d+',regex.search(time).group()).group())
        return hour
    
```

```python
convertime(df['usetime'][0])
```

```python
df['usetime'] = df['usetime'].apply(convertime)
```

```python
df
```

```python
# 흡입력 단위 맞춰주기 및 숫자형태로 변경
def suction(suc):
    if suc:
        if ('W' or 'AW') in suc:
            return int(re.search('\d+',suc).group())
        elif 'Pa' in suc:
            suc = suc.replace(',','')
            return int(re.search('\d+',suc).group())/100
suction(df['suction'][0])
```

```python
df['suction'] = df['suction'].apply(suction)
```

```python
df
```

```python
# 필요없는열 버리기(근데 이거 안하고 뒤에 재정렬하면서 그냥 버려도 무관)
df.drop(['spec_lst','title'], axis = 1, inplace=True)
```

```python
# 재정렬
df = df[['product','company','price','usetime','suction','spec']]
df
```
