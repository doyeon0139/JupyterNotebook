```python
df = pd.read_csv('./data_t/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20211031.csv', encoding='euckr')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df['분양가격(제곱미터)'].dtype
```

```python
df['분양가격(제곱미터)'].astype('float')
```

```python
errors = []
for i, j in enumerate(df['분양가격(제곱미터)']):
    try:
        float(j)
    except:
        errors.append(j)
```

```python
errors
```

```python
df['분양가격(제곱미터)'] = df['분양가격(제곱미터)'].apply(lambda x: str(x).strip().replace(',', ''))
```

```python
idexes = []
errors = []
for i, j in enumerate(df['분양가격(제곱미터)']):
    try:
        float(j)
        idexes.append(i)
    except:
        errors.append(j)
```

```python
errors
```

```python
idexes
```

```python
df_clean = df.iloc[idexes]
```

```python
df_clean['분양가격(제곱미터)'] = df_clean['분양가격(제곱미터)'].astype('float')
```

```python
df_clean['평당분양가'] = df_clean['분양가격(제곱미터)'] * 3.3
df_clean
```

```python
df_clean['규모구분'].unique()
```

```python
df_clean.info()
```

```python
# df_clean['규모구분'] = df_clean['규모구분']
df_clean['규모구분'] = df_clean['규모구분'].str.replace('전용면적','')
df_clean['규모구분'] = df_clean['규모구분'].str.replace('제곱미터','')
df_clean['규모구분'] = df_clean['규모구분'].str.replace('이하','')
df_clean['규모구분'] = df_clean['규모구분'].str.replace('초과','~')
df_clean['규모구분']
```

```python
df_clean.drop('분양가격(제곱미터)', axis=1, inplace=True)
```

```python
df_clean.head()
```

```python
dfc = df_clean.copy()
```
