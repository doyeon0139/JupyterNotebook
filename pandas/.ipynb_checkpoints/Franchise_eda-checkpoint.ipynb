```python
import folium
```

```python
lat = df_con['위도'].mean()
long = df_con['경도'].mean()
```

```python
folium.Map([lat, long])
```

```python
m = folium.Map([lat, long], zoom_start=12)

for i in df_con.index:
    sub_lat = df_con.loc[i, '위도']
    sub_long = df_con.loc[i, '경도']
    title = df_con.loc[i, '상호이름']
    
    folium.Marker([sub_lat, sub_long],
                 tooltip=title).add_to(m)
    
m
```
