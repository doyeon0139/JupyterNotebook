```python
pp['공원보유시설(운동시설)'].values
```

```python
gym = pp1['공원보유시설(운동시설)'].unique().tolist()
gym = str(gym)
gym
```

```python
import re
```

```python
gym = gym.replace('+','').replace("'", '.').replace(" ", '.').replace("/", '.')
gym
```

```python
re.sub('\d', '', gym)
gym
```

```python
gym = re.sub("[\[\d\-nan\.\]]","",gym)
gym
```

```python
gym = re.sub(".{2.}", ".", gym)
gym
```

```python
pd.Series(gym.split(',')).value_counts().iloc[:30].plot(kind='bar', figsize = (15,5))
```
