```python
# europe 또는 aisa 국가만 출력하고, 각각 인구수가 가장 많은 국가를 출력하시오..
```

```python
# country_df[(country_df.Continent == 'Asia') | (country_df.Continent == 'Europe')]
```

```python
sql = """
select * from country where population in (select max(population) \
from country where continent = 'asia' or continent = 'europe' group by continent);
"""
pd.read_sql(sql, db)
```

```python
sql = """
select * from rental;
"""
rental_df = pd.read_sql(sql, db)
rental_df.head(3)
```

```python
# rental_df['month'] = rental_df.rental_date.dt.month
rental_df['month'] = rental_df.rental_date.astype('str').apply(lambda x : x[:7])
```

```python
rental_df.groupby('month').rental_id.count()
```

```python
sql = """
select rental_id, left(rental_date, 7) as 'month', count(*) as 'rental_count'
    from rental
    group by month;
"""
rental_df = pd.read_sql(sql, db)
rental_df
```

```python
# 월별 총매출, 평균매출, 렌탈건수 합 을 출력하시오..
```

```python
sql = """
select * from payment;
"""
payment_df = pd.read_sql(sql, db)
payment_df.info()
```

```python
payment_df['month'] = payment_df.payment_date.dt.month

grouped = payment_df.groupby('month')
```

```python
payment_df.head(2)
```

```python
grouped.amount.sum()
```

```python
grouped.amount.mean()
```

```python
grouped.amount.count()
```

```python
# groupby 객체 메소드 : agg, filter, transform

# agg : groupby 객체에 함수를 적용하는 메소드
# filter : 특정 group을 추출
# transform : 그룹의 값으로 데이터값을 바꿈
```

```python
payment_df.groupby('month').agg({'payment_id': np.size, 'amount': [np.sum, np.mean]})
```

```python
sql = """
select count(payment_id) '구매건수', avg(amount) '평균구매액', sum(amount) '총구매액',
       left(payment_date, 7) as month
        from payment
        group by month;
"""
payment_df = pd.read_sql(sql, db)
payment_df
```
