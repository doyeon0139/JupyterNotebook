```python
# groupby
# 동일한 기준을 갖는 데이터끼리 데이터를 그룹으로 묶는다

# 1. 데이터를 묶는 과정 => group생성
# 2. group별 어떤 값을 출력할지 => group별 연산과정
```

```python
df = pd.read_csv('./data/gapminder.tsv', sep='\t')
```

```python
df.head()
```

```python
df.groupby('year')['lifeExp'].mean()
```

```python
grouped = df.groupby('year')
grouped.lifeExp.mean()
```

```python
# grouped 객체에 적용되는 메소드

# agg - 함수 적용하는 메소드, 사용자 정의함수를 적용시
# filter - 원하는 그룹만 추출
# transform - 데이터의 값을 그룹의 값으로 변경
```

```python
# agg - 사용자정의함수, 다양한 함수를 적용
```

```python
grouped = df.groupby('year')
grouped.lifeExp.agg(np.mean)
```

```python
def mean_dff(values, diff_value):
    n=len(values)
    suma = 0
    for value in values:
        suma += value
    mean = suma/n
    return mean - diff_value       # 평균과의 차이를 계산
```

```python
global_mean = df.lifeExp.mean()     # 데이터의 평균
```

```python
df.groupby('year').lifeExp.agg(mean_dff, diff_value = global_mean)    # 편차 계산
```

```python
grouped = df.groupby('year')
grouped.agg([np.mean, np.std, np.count_nonzero])
```

```python
df.groupby('year').lifeExp.agg([np.mean, np.std, np.count_nonzero])
```

```python
df.groupby('year').agg({'lifeExp':'mean','pop':'median','gdpPercap':'max'})
```
