```python
# 월별 총매출, 평균매출, 렌탈건수 합을 출력
```

```python
sql = """
select * from payment;
"""

tt = pd.read_sql(sql, db)
tt
```

```python
tt['month'] = tt['payment_date'].dt.month
```

```python
tt.head()
```

```python
grouped = tt.groupby('month')
```

```python
grouped.amount.sum()
```

```python
grouped.amount.mean()
```

```python
grouped.amount.count()
```

```python
# groupby 객체 메소드 : agg, filter, transform
# agg : groupby 객체에 함수를 적용하는 메소드
# filter : 특정 group을 추출
# transform : 그룹의 값으로 데이터값을 바꿈
```

```python
tt.groupby('month').agg({'payment_id': np.size, 'amount':[np.sum, np.mean]})    #     *agg함수?????!!!!!
```

```python
sql = """
select count(payment_id) '구매건수', avg(amount) '평균구매액', sum(amount) '총구매액', left(payment_date, 7) as 'month'
from payment
group by month;
"""

tt = pd.read_sql(sql, db)
tt
```
