```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./ml_source_files/USA_Housing.csv')
```

```python
df.head()
```

```python
# 집값을 예측하는 모델

# y : 집값(price) - continuous data => regression
# x : income, house age, nums of rooms, nums of bedrooms
```

```python
y = df['Price']
x = df.drop(['Price', 'Address'], axis=1)      # 집값에 필요없는 것들
```

```python
# train test split

# train : 모델을 훈련하는데 사용하는 데이터
# test :  모델을 평가하는데 사용하는 데이터
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
#algorithm
```

```python
lr = LinearRegression()
```

```python
# model train
```

```python
lr.fit(x_train, y_train)
```

```python
# predict
```

```python
preds = lr.predict(x_test)
```

```python
preds
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(preds, y_test)
```
