```python
from sklearn.datasets import make_blobs

x,y = make_blobs(n_samples= 150, n_features=2, centers= 3, cluster_std= 0.5, shuffle=True, random_state=0)
```

```python
plt.scatter(x[:,0], x[:,1], c = 'white', marker='o', edgecolors='black', s=50)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2, init='random', max_iter= 300, random_state=111)

y_km = km.fit_predict(x)
y_km
```

```python
km.cluster_centers_
```

```python
plt.scatter(x[y_km==0, 0], x[y_km==0, 1], s=50, c='lightgreen', marker='s', edgecolors='black', label='cis 1')

plt.scatter(x[y_km==1, 0], x[y_km==1, 1], s=50, c='orange', marker='o', edgecolors='black', label='cis 2')

plt.scatter(x[y_km==0, 0], x[y_km==0, 1], s=50, c='lightblue', marker='v', edgecolors='black', label='cis 3')

plt.scatter(x[y_km==0, 0], x[y_km==0, 1], km.cluster_centers_[:,1], s =250, c='lightblue', marker='v', edgecolors='black', 
            label='cis 2')
```

```python
# silhouette
# 군집 내의 데이터들이 얼마나 결합력있게 그룹핑이 되어있는지(뭉쳐있는지)
#  군집내의 샘플 간의 거리 - a
# 군집 간의 샘플간의 거리 계산 - b
# 군집 내의 데이터는 가까울수록 좋고, 군집 간의 데이터의 거리는 멀수록 좋다
# silhouette = (b-a)/max(a,b)
```

```python
from matplotlib.pyplot import cm
from sklearn.metrics import silhouette_samples
```

```python
km = KMeans(n_clusters=3, 
            init='k-means++', 
            random_state=111)
y_km = km.fit_predict(x)

silhouette_vals = silhouette_samples(x, y_km, metric='euclidean')
silhouette_vals
```

```python
# silhouette이 가장 정확한 cluster 평가 지표
# 그러나 cluster의 절대적인 평가지표는 존재하지 않는다
# 시각화를 통한 silhouett 평가
# cluster 별로 silhouette 점수가 고른지 여부 판단 - cluster의 쏠림 여부
# silhoueete 평균점수값

y_ax_lower, y_ax_upper = 0, 0
y_ticks = []
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]

clist = ['g','b','y']
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km==c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0,
            edgecolor = 'none', color=clist[i])
#     y_ticks.append(y_ax_lower + y_ax_upper)/2)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color='red', linestyle='--')

plt.show()
```

```python
km = KMeans(n_clusters=2, 
            init='k-means++', 
            random_state=111)
y_km = km.fit_predict(x)

silhouette_vals = silhouette_samples(x, y_km, metric='euclidean')
silhouette_vals

y_ax_lower, y_ax_upper = 0, 0
# y_ticks = []
cluster_labels = np.unique(y_km)  # 0, 1, 2
n_clusters = cluster_labels.shape[0]  # 3

clist = ['g','b','y','o','p','b']
# clist=[0,1,2,3,4]


for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km==c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0,
            edgecolor = 'none', color=clist[i])
#     y_ticks.append(y_ax_lower + y_ax_upper)/2)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color='red', linestyle='--')

plt.show()
```

### Hierachical Clustering
* 모든 샘플들의 거리행렬을 계산
* 가장 멀리이쓴 개체들의 거리에 기반하여 가장 가까운 두 개의 군집을 합한다
* 거리행렬을 다시 계산한다 
* 하나의 군집이 남을 때까지 반복한다
* dendrogram을 드릴 수 있다.

### DBSCAN (밀도기반 클러스터링)
* 밀도 - 특정 반지름(e) 내에 점(데이터)들의 수
* 특정 데이터를 중심으로 반지름(e) 내에 이웃점의 갯수가 설정한 수(MinPts)보다 많으면 중심점이 된다
* 경계점 - e 내에서 중심점은 아니지만 다른 중심점의 e 범위 내에 존재하는 경우
* 잡음점 - 중심점도 경계점도 아닌 다른 모든 점
* 중심점이 서로 다른 군집의 일부가 겹치면 그 군집은 서로 연결되어 하나의 군집이 된다
* 군집의 모양에 대한 가정이 불필요하다
* 잡음의 제거가 쉽다
* 차원이 증가하면 부정적 영향

```python
from sklearn.datasets import make_blobs

x,y = make_blobs(n_samples= 150, n_features=2, centers= 3, cluster_std= 0.7, shuffle=True, random_state=0)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics.cluster import silhouette_score

db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db = db.fit_predict(x)
# y_db
```

```python
silhouette_score(x, y_db)
```

```python
km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=300, random_state=111)
y_km = km.fit_predict(x)
```

```python
agg = AgglomerativeClustering(n_clusters=3)
y_agg = agg.fit_predict(x)
```

```python
print('km_sil_score : ', silhouette_score(x, y_km, metric='euclidean'))
print('db_sil_score : ', silhouette_score(x, y_db, metric='euclidean'))
print('agg_sil_score : ', silhouette_score(x, y_agg, metric='euclidean'))
```

```python
from sklearn.datasets import make_moons

x,y = make_moons(n_samples=200, noise=0.05, random_state=111)
plt.scatter(x[:,0], x[:,1])
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics.cluster import silhouette_score

db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db = db.fit_predict(x)
```

```python
km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=300, random_state=111)
y_km = km.fit_predict(x)
```

```python
agg = AgglomerativeClustering(n_clusters=3)
y_agg = agg.fit_predict(x)
```

```python
print('km_sil_score : ', silhouette_score(x, y_km, metric='euclidean'))
print('db_sil_score : ', silhouette_score(x, y_db, metric='euclidean'))
print('agg_sil_score : ', silhouette_score(x, y_agg, metric='euclidean'))
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import silhouette_score
scaler = StandardScaler()     # 데이터 정규화(standardization)  =  (x - x_mean)/x_std
scaler = MinMaxScaler()  # 데이터 표준화 = (x - x_min)/(x_max - x_min)
scaler.fit(x)
x_scaled = scaler.transform(x)
x_scaled
```

```python
fig, axes = plt.subplots(1,4, figsize=(15,3))

random_state = np.random.RandomState(seed=0)
random_clusters = random_state.randint(low=0, high=2, size=len(x))
random_clusters

axes[0].scatter(x_scaled[:,0], x_scaled[:,1], c=random_clusters, cmap=mglearn.cm3, s=10, edgecolor='black')
axes[0].set_title('무작위 할당 : {}'.format(silhouette_score(x_scaled, random_clusters)))

algorithms = [KMeans(n_clusters=2), AgglomerativeClustering(n_clusters=2), DBSCAN()]


for ax, algorithm in zip(axes[1:], algorithms):
    clusters = algorithm.fit_predict(x_scaled)
    ax.scatter(x_scaled[:,0], x_scaled[:,1], c=clusters, cmap=mglearn.cm3, s=60, edgecolor='black')
    ax.set_title("{} : {}".format(algorithm.__class__.__name__, round(silhouette_score(x_scaled, clusters),2)))
```
