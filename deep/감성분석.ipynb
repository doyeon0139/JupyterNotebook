```python
train.dropna(inplace=True)          # nan값 제거
test.dropna(inplace=True)
```

```python
from konlpy.tag import Komoran
```

```python
kor = Komoran()
```

```python
def text_clean(doc):
    text = re.sub('[^가-핳ㄱ-ㅎㅏ-ㅣ ]','', doc)   # 한글이 아닌것을 제외 정규표현식
    return text
```

```python
train['document1'] = train.document.apply(text_clean)
test['document1'] = test.document.apply(text_clean)
```

```python
train.head()
```

```python
test.head()
```

```python
# stopwords-ko.txt
# 불용어 : 텍스트를 분석할 때 의미를 가지지 않는 형태소..
```

```python
stw = pd.read_csv('stopwords-ko.txt', header=None)
stwords = list(stw.iloc[:,0].values)
stwords
```

```python
def remove_stopwords(doc):
    text = [w for w in doc.split() if w not in stwords]
    return ' '.join(text)
```

```python
train['document2'] = train.document1.apply(remove_stopwords)  # 불용어 제거
```

```python
test['document2'] = test.document1.apply(remove_stopwords)
```

```python
train.head()
```

```python
test.head()
```

```python
train = train[train.document2 != '']
test = test[test.document2 != '']
```

```python
train_docs = []

for d in train['document2']:
    tokens = kor.nouns(d)    # 명사만 추출
    train_docs.append(tokens)
```

```python
test_docs = []

for d in test['document2']:
    tokens = kor.nouns(d)
    test_docs.append(tokens)
```

```python
train_docs
```

```python
for i in range(len(train_docs)):     # 리스트를 문자열로 바꿈
    train_docs[i] = ' '.join(train_docs[i])
```

```python
for i in range(len(test_docs)):
    test_docs[i] = ' '.join(test_docs[i])
```

```python
train['train_docs'] = train_docs
test['test_docs'] = test_docs
```

```python
train
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vec = CountVectorizer(min_df=3, max_df=0.9, max_features=2000)  
# 90% 이상 등장은 제외, 3번이하 token들 제거(사람이름같은거 제외)/실수로 주면 비율 , feature 2000개 까지만
x_train = vec.fit_transform(train_docs)
x_train
```

```python
vec = CountVectorizer(min_df=3, max_df=0.9, max_features=2000) 
x_test = vec.fit_transform(test_docs)
x_test
```

```python
x_train = x_train.toarray()
x_test = x_test.toarray()
```

```python
y_train = train.label
```

```python
y_test = test.label
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
preds = log.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, preds))
```
