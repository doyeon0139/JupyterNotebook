```python
news_df = pd.read_csv('news_df.csv', index_col=0)
news_df
```

```python
def pre(text):
    text = re.sub('[^가-힣ㄱ-ㅎ ]', ' ', text[0])
    return text
```

```python
news_df['news1'] = news_df.news.apply(lambda x : pre(x))
```

```python
stw = pd.read_csv('stopwords-ko.txt', header=None)
stopwords = list(stw.iloc[:, 0].values)
stopwords

def remove_stopwords(doc):
    text = [w for w in doc.split() if w not in stopwords]
    return ' '.join(text)
```
