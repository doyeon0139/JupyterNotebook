```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.layers import Dense, Dropout, Activation, Embedding, SimpleRNN, LSTM
from tensorflow.keras.models import Sequential, load_model, Model
from tensorflow.keras.utils import to_categorical
from tensorflow import keras
from tensorflow.keras.preprocessing import sequence
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import re
```

```python
with open('./deep_data/lastofus.txt', encoding='euckr') as f:        # 파일불러오기
    line = f.readlines()
    
    for i in line:
        print(i)
```

```python
text = ''

array = []

with open('./deep_data/lastofus.txt', encoding='euckr') as f:
    lines = f.readlines()
    text = text.join([i for i in lines if re.match(r'^[A-Z].*:', i)])   # 대화체 문장만 가져오기(:으로 끝나는 문장을 가져감)

text
```

```python
print(text)
```

```python
arr = []
for t in text.split('\n')[:1001]:            # 문장들을 줄바꿈단위로 한문장씩 arr에 담는다
    arr.append(t)
```

```python
arr
```

```python
token = Tokenizer(lower=False, filters='.,?:;\'\"-!%\n')  # 특수기호 빼고 토큰화선언  / 단어단위로 만들기
token.fit_on_texts(arr)  #  arr를 토큰화 해서 단어사전을 생성한다
```

```python
token.word_index
```

```python
len(token.word_index)
```

```python
seq = token.texts_to_sequences(arr)    # 토큰화 된 단어들을 숫자값으로 단어를 바꿈
```

```python
lenghts = [len(line) for line in seq]
lenghts
```

```python
plt.hist(lenghts, bins=50)
```

```python
seq
```

```python
seq = pad_sequences(seq, maxlen=12)     # pad_sequences를 통해 모든 샘플의 길이를 동일하게 맞춤(12), 
seq.shape                               # 12보다 짧은 길이의 문장은 0으로 채운다
```

```python
x = seq[:-1]   # x,y나누기
y = seq[1:]
```

```python
x.shape
```

```python
y.shape
```

```python
x = x.reshape(1000, 12, 1)
```

```python
y = y.reshape(1000, 12, 1)
```

```python
from tensorflow.keras.utils import to_categorical
y_oh = to_categorical(y)     # output이 문장이기 때문에 category로 만들어줌 
```

```python
y_oh.shape
```

```python
model=Sequential()
model.add(Embedding)
model.add(LSTM(128, input_shape=(12,1), return_sequences=True)) # return_sequences : 각 hidden state들이 output을 보낸다
model.add(LSTM(256, return_sequences=True))  

model.add(Dense(2468, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(x, y_oh, epochs=20, batch_size=20)
```

```python
pre = model.predict(x[7:8])  # x의 7번째 문장을 예측
```

```python
idx_word = {}

for w in token.word_index:                       # 문장복구 하기위해 딕셔너리에 넣기
      idx_word[token.word_index[w]] = w

idx_word
```

```python
input_x = x[30:31]           # 입력값

input_txt = [ t[0] for t in input_x[0]]      # 문장복구

temp = ''

for t in input_txt:
      if t != 0:
        temp += idx_word[t]
        temp += ' '
temp       # 입력값의 문장
```

```python
pred = model.predict(input_x)         # 모델이 예측한 문장
pred_tok = np.argmax(pred, axis=2)

temp = ''

for l in pred_tok:
      for w in l:
        if w != 0:
          temp += idx_word[w]
          temp += ' '
temp
```
