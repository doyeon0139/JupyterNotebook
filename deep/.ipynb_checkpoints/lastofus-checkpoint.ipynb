```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.layers import Dense, Dropout, Activation, Embedding, SimpleRNN, LSTM
from tensorflow.keras.models import Sequential, load_model, Model
from tensorflow.keras.utils import to_categorical
from tensorflow import keras
from tensorflow.keras.preprocessing import sequence
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import re
```

```python
with open('./deep_data/lastofus.txt', encoding='euckr') as f:        # 
    line = f.readlines()
    
    for i in line:
        print(i)
```

```python
text = ''

array = []

with open('./deep_data/lastofus.txt', encoding='euckr') as f:
    lines = f.readlines()
    text = text.join([i for i in lines if re.match(r'^[A-Z].*:', i)])   #   (:   )

text
```

```python
print(text)
```

```python
arr = []
for t in text.split('\n')[:1001]:            #    arr 
    arr.append(t)
```

```python
arr
```

```python
token = Tokenizer(lower=False, filters='.,?:;\'\"-!%\n')  #     /  
token.fit_on_texts(arr)  #  arr    
```

```python
token.word_index
```

```python
len(token.word_index)
```

```python
seq = token.texts_to_sequences(arr)    #      
```

```python
lenghts = [len(line) for line in seq]
lenghts
```

```python
plt.hist(lenghts, bins=50)
```

```python
seq
```

```python
seq = pad_sequences(seq, maxlen=12)     # pad_sequences      (12), 
seq.shape                               # 12    0 
```

```python
x = seq[:-1]   # x,y
y = seq[1:]
```

```python
x.shape
```

```python
y.shape
```

```python
x = x.reshape(1000, 12, 1)
```

```python
y = y.reshape(1000, 12, 1)
```

```python
from tensorflow.keras.utils import to_categorical
y_oh = to_categorical(y)     # output   category  
```

```python
y_oh.shape
```

```python
model=Sequential()
model.add(Embedding)
model.add(LSTM(128, input_shape=(12,1), return_sequences=True)) # return_sequences :  hidden state output 
model.add(LSTM(256, return_sequences=True))  

model.add(Dense(2468, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(x, y_oh, epochs=20, batch_size=20)
```

```python
pre = model.predict(x[7:8])  # x 7  
```

```python
idx_word = {}

for w in token.word_index:                       #    
      idx_word[token.word_index[w]] = w

idx_word
```

```python
input_x = x[30:31]           # 

input_txt = [ t[0] for t in input_x[0]]      # 

temp = ''

for t in input_txt:
      if t != 0:
        temp += idx_word[t]
        temp += ' '
temp       #  
```

```python
pred = model.predict(input_x)         #   
pred_tok = np.argmax(pred, axis=2)

temp = ''

for l in pred_tok:
      for w in l:
        if w != 0:
          temp += idx_word[w]
          temp += ' '
temp
```
