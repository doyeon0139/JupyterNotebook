```python
train_data = pd.read_table('./deep_data/ratings_train.txt')
```

```python
train_data = train_data.dropna()
```

```python
train_data['clean_doc'] = train_data['document'].str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣]A-Za-z', ' ')
```

```python
train_data
```

```python
stw = pd.read_csv('./deep_data/stopwords-ko.txt', header=None)
stw_li = stw[0].values
```

```python
from konlpy.tag import Mecab
```

```python
mec = Mecab(dicpath='C:/mecab/mecab-ko-dic/')
tokenized_data = []
for sentence in train_data['clean_doc']:
    temp_x = mec.morphs(sentence)
    temp_x = [word for word in temp_x if not word in stw_li]
    tokenized_data.append(temp_x)
```

```python
tokenized_data
```

```python
token_count={}
for line in tokenized_data:
    for tok in line:
        if tok in token_count:
            token_count[tok] += 1
        else:
            token_count[tok] = 1
token_count
```

```python
len(token_count)
```

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences=tokenized_data, vector_size=100, window=5, min_count=5, sg=0)
# snetence:token데이터, size 몇차원, window 주변 무라고? 5, min_count: 최소 5번 등장한것을 벡터라이징 하겠다, sg는?
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentences = [
    'nice great best amazing',
    'sop lies',
    'pitiful nerd',
    'supreeme quality',
    'bad',
    'highly repectable'
]
y_train = [1,0,0,1,1,0]
```

```python
t = Tokenizer()
t.fit_on_texts(sentences)
vocab_size = len(t.word_index)+1

print(vocab_size)
```

```python
t.word_index
```

```python
x_encoded = t.texts_to_sequences(sentences)
```

```python
x_train = pad_sequences(x_encoded, maxlen=4)
y_train = np.array(y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten

model = Sequential()
model.add(Embedding(vocab_size, 4, input_lenght=4))
```

```python
import tensorflow_hub as hub
import tensorflow as tf
```

```python
embedding = 'https://tfhub.dev/google/tf2-preview/gnews-swivel-20dim/1'
hub_layer = hub.KerasLayer(embedding, input_shape=[], dtype=tf.string, trainable=True)
```

```python
import gensim.downloader as api
path = api.load('word2vec-google-news-300', return_path=True)
path
```

```python
hub_layer.shape
```

```python
model = Sequential()
model.add(hub_layer)
model.add(Dense(14, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
model.fit(x_train, y_train, epochs=100, verbose=2)
```
