```python
import pandas as pd
import numpy as np
import re
import time

import requests
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import warnings
warnings.filterwarnings(action='ignore')
```

```python
driver = webdriver.Chrome()
```

```python
title_li = []
news_li = []
category = ['경제','사회','생활문화']
cate_li = []

url = 'https://news.naver.com/'
elem = driver.get(url)


for i,ca in enumerate(range(3,6)):
    url = 'https://news.naver.com/'
    elem = driver.get(url)
    
    driver.find_element_by_css_selector(f'body > section > header > div.Nlnb._float_lnb > div > div > div.Nlnb_left._lnb_scroll > div > div > ul > li:nth-child({ca}) > a').click()
    
    
    for news in range(1,6):
        cate_li.append(category[i])
        driver.find_element_by_css_selector(f"#main_content > div > div._persist > div:nth-child(1) > div:nth-child({news}) > div.cluster_body > ul > li:nth-child(1) > div.cluster_text > a").click()
        time.sleep(1)

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        path_a = '#main_content'
        sell = soup.select(path_a)[0]

        # 제목
        title = sell.select("#articleTitle")[0].text
        title_li.append(title)

        # 본문
        news = sell.select('#articleBodyContents')[0].text
        news_li.append(news)

        driver.find_element_by_css_selector(f'#lnb > ul > li:nth-child({ca})').click()
```

```python
title_li
```

```python
news_df = pd.DataFrame()
news_df['category'] = cate_li
news_df['title'] = title_li
news_df['document'] = news_li
news_df
```

```python
news_df.drop('title', axis=1, inplace=True)
```

```python
news_df
```

```python
def text_clean(doc):                    
    text = re.sub('[^가-핳ㄱ-ㅎㅏ-ㅣ ]','', doc)  
    return text
```

```python
news_df['document'] = news_df.document.apply(text_clean)
news_df
```

```python
stw = pd.read_csv('stopwords-ko.txt', header=None)       
stopwords = list(stw.iloc[:, 0].values)
# stopwords

def remove_stopwords(doc):
    text = [w for w in doc.split() if w not in stopwords]
    return ' '.join(text)
```

```python
news_df['news'] = news_df.document.apply(remove_stopwords)
```

```python
news_df
```

```python
from konlpy.tag import Hannanum, Komoran, Kkma
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN
```

```python
kor = Komoran()
```

```python
docs = []

for n in news_df['news']:
    docs.append(kor.nouns(n))
    
docs
```

```python
for i in range(len(docs)):
    docs[i] = ' '.join(docs[i])
    
docs
```

```python
vec = CountVectorizer(max_df=0.7, min_df=2)

x = vec.fit_transform(docs)
x
```

```python
news_arr = x.toarray()
news_arr
```

```python
km = KMeans(n_clusters=3)
dbs = DBSCAN(eps=50, min_samples=5)
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
```

```python
km.fit_predict(news_arr)
```

```python
dbs.fit_predict(news_arr)
```

```python
agg.fit_predict(news_arr)
```

```python
vec = TfidfVectorizer(max_df=0.7, min_df=2, ngram_range=(1, 2))

x = vec.fit_transform(docs)
x.toarray()
```
