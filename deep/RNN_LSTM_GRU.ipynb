```python
st_df = pd.read_csv('./deep_data/stopwords-ko.txt', header=None)
st_li = [st[0] for st in st_df.values]

def remove_stwords_onletter(doc):
    doc = mec.morphs(doc)
    doc = [t for t in doc if t not in st_li]
    doc = [t for t in doc if len(t) > 1]
    doc = ' '.join(doc)
    return doc
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Bidirectional, LSTM, GRU

from tensorflow.keras.datasets import reuters
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model
```

```python
from tensorflow.keras.layers import Conv1D, GlobalAveragePooling1D, Dropout, Flatten, Input, Concatenate
from tensorflow.keras import Model
```

```python
vocal_size = 5000
max_len = 30
embedding_dim = 128
dropout_prob = (0.5, 0.8)
num_filters = 128

```

```python
model_input = Input(shape = (max_len,))
z = Embedding(vocab_size, embedding_dim, input_length = max_len)(model_input)

conv_blocks = []

for sz in [3,4,5]:
    conv = Conv1D(filters = num_filters,
                 kernel_size = sz,
                 padding = 'valid',
                 activation='relu',
                 strides=1)(z)
    conv = GlobalAveragePooling1D()(conv)
    conv = Flatten()(conv)
    conv_blocks.append(conv)
    
z = Concatenate()(conv_blocks) if len(conv_blocks) > 1 else conv_blocks[0]   # 이어붙이기
z = Dropout(dropout_prob[1])(z)
model_output = Dense(1, activation='sigmoid')(z)

model = Model(model_input, model_output)
```

```python
model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', mode = 'min', verbose=1, patience=5)
mc = ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
model.fit(x_train, y_train, epochs=10, callbacks=[es, mc], batch_size=64, validation_data=(x_test, y_test))
```
