```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('./deep_data/Concrete_Data.xls')
df.head()
```

```python
df.info()
```

```python
df.columns = ['cement', 'furnace','ash','water','superplasticizer','coarse','fine','age','strength']
```

```python
df.head()
```

```python
df.info()
```

```python
y = df.values[:,8]
x = df.values[:,:8]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_sc = sc.transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x.shape
```

```python
model = Sequential()
model.add(Dense(30, input_dim=8, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
hist = model.fit(x_train_sc, y_train, validation_split=0.2, epochs=200, batch_size=20)
```

```python
model.evaluate(x_test_sc, y_test)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
model = Sequential()
model.add(Dense(256, input_dim=8, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(56, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')

hist = model.fit(x_train_sc, y_train, validation_split=0.2, epochs=200, batch_size=20)

model.evaluate(x_test_sc, y_test)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
pred = model.predict(x_test_sc)
```

```python
pred = [v[0] for v in pred]
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```
