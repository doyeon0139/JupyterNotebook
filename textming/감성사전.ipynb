```python
from nltk.corpus import stopwords
```

```python
stopwords = stopwords.words('english')
```

```python
tokenizer = RegexpTokenizer('[\w]+')   # 정규표현식 단어단위로 토큰자르기
```

```python
p_stemmer = PorterStemmer()
```

```python
raw = lines1.lower()
tokens = tokenizer.tokenize(raw)
stopped_tokens = [i for i in tokens if not i in stopwords]
stopped_tokens
```

```python
stop_tokens
```

```python
match = [x for x in stopped_tokens if x in list(nrc[0])]
```

```python
emotion = []

for i in match:
    em_word = nrc[nrc[0] == i][1].values
    for j in em_word:
        emotion.append(j)
```

```python
sent_result = pd.Series(emotion).value_counts()
sent_result
```

```python
# 점수계산
def emotion_score(series):
    emotion_score = 0
    for i in range(len(series)):
        if series.index[i] in ('positive','anticipation','trust','joy','surprise'):
            emotion_score += series.iloc[i]
            
        else:
            emotion_score -= series.iloc[i]
            
    return emotion_score
```

```python
emotion_score(sent_result)
```
