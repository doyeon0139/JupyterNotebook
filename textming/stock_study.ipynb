```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import requests
from tensorflow.keras.layers import *
from tensorflow.keras.models import * 
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler

```

```python
import FinanceDataReader as fdr
import PIL
import PIL.Image
```

```python
df = fdr.DataReader('KS11', '2018-01-01', '2022-03-20')
df.tail()
```

```python
df.head(2)
```

```python
scaler = MinMaxScaler()              # 금액들을 scaler 
price = scaler.fit_transform(df)
price[:3]
```

```python
window_size = 10
X = []
Y = []

for i in range(len(price) - window_size):                 # 10일전 예측
  X.append([price[i+j] for j in range(window_size)])
  Y.append(price[window_size + i][0])
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
X.shape
```

```python
Y.shape
```

```python
tt_split = 800                 # train, test 나누기

x_train = X[:tt_split]
y_train = Y[:tt_split]

x_test = X[tt_split:]
y_test = Y[tt_split:]
```

```python
# 3차원입력 : (n_sample, time_steps, n_features)

x_train.shape
```

```python
PIL.Image.open('./deep_data/lstm_img.jpg')
```

```python
model = Sequential()
model.add(LSTM(100, input_shape=(10, 6)))                # output을 100개의 원소를 가진 vector
model.add(Dropout(0.2))                                  # 네트웍의 20% drop
model.add(Dense(1, activation='linear'))                 
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(x_train, y_train, epochs=20, batch_size=10, validation_data=(x_test, y_test))
```

```python
x_train_pred = model.predict(x_train)           # 예측하기
x_test_pred = model.predict(x_test)
```
