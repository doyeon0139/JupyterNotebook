* 영업이익률 = 영업이익/매출액  
* 자기자본이익률(ROE) = 순이익/자기자본  
* 총자본이익률(ROA) = 순이익/총자산  
  
* 유동비율 = 유동자산/유동부채  
* 이자보상비율 = 영업이익/이자비용  
  
* 주당순이익(EPS) = 순이익/주식수
* PER = 주가/EPS
* 주당순자산(BPS) = 자기자본/주식수
* PBR = 주가/BPS  
  
* 매출성장률
* 영업이익성장률

### 연도별 roa상위 10 종목에 투자할 경우 투자수익률 구하기

```python
yearly_price_df = df.pivot(index='year', columns='Name', values='수정주가')
yearly_price_df
```

```python
yearly_rtn_df = yearly_price_df.pct_change().shift(-1)
yearly_rtn_df
```

```python
yearly_rtn_df['AD모터스']
```

```python
indicator = 'ROA'
top_n = 10

df.groupby(['year'])[indicator].nlargest(top_n)
```

```python
top_n_indicator_df = df.groupby(['year'])[indicator].nlargest(top_n).reset_index()
```

```python
top_n_indicator_df
```

```python
df.head()
```

```python
df.loc[1079]
```

```python
top_n_roa_df = df.loc[top_n_indicator_df['level_1']]     #?????????
top_n_roa_df
```

```python
indicator_df = top_n_roa_df.pivot(index='year', columns='Name', values='ROA')
indicator_df.head()
```

```python
asset_on_df = indicator_df.notna().astype(int).replace(0, np.nan)
asset_on_df
```

```python
asset_on_df.notna().sum(axis=1)
```

```python
selected_return_df = yearly_rtn_df * asset_on_df
selected_return_df
```

```python
selected_return_df.notna().sum(axis=1)
```

```python
rtn_series = selected_return_df.mean(axis=1)
rtn_series      # 연도별 투자 수익률
```

```python
(rtn_series + 1).cumprod().plot()
```

### 소형주 + 저pbr 전략

```python
marker_cap_quantile_series = df.groupby('year')['시가총액'].quantile(0.2)   # 시가총액 하위 20
```

```python
marker_cap_quantile_series
```

```python
filtered_df = df.join(marker_cap_quantile_series, on='year', how='left', rsuffix='qunatile_0.2')   # ?????
filtered_df
```

```python
filtered_df = filtered_df[filtered_df['시가총액'] <= filtered_df['시가총액qunatile_0.2']]  # 소형주
```

```python
filtered_df = filtered_df[filtered_df['pbr'] >= 0.2]
```

```python
smallest_pbr_series = filtered_df.groupby('year')['pbr'].nsmallest(15)
smallest_pbr_series
```

```python
selected_index = smallest_pbr_series.index.get_level_values(1)
selected_df = filtered_df.loc[selected_index]
selected_df
```

```python
selected_df = selected_df.pivot(index='year', columns='Name', values='pbr')
```

```python
selected_df.head()
```

```python
asset_on_df = selected_df.notna().astype(int).replace(0, np.nan)
asset_on_df
```

```python
selected_rtn_df = asset_on_df * yearly_rtn_df
selected_rtn_df  # 연도별 수익률
```

```python
selected_rtn_df.mean(axis=1) # 누적수익률
```

```python
cum_ftn_str_df = (selected_rtn_df.mean(axis=1) + 1).cumprod()
```

```python
cum_ftn_str_df.plot(figsize=(15,3))
```

### 벤저민 그레엄

* roa 5% 이상
* 부채비율 50% 이하
* pbr 0.2 이상
* pbr 낮은 순으로 20~30개 매수

```python
filtered_df = df[df['ROA'] >= 0.05]
filtered_df['부채비율'] = filtered_df['비유동부채']/filtered_df['자산총계']
filtered_df = filtered_df[filtered_df['부채비율'] <= 0.5]
filtered_df = filtered_df[filtered_df['pbr'] >= 0.2]
```

```python
filtered_df
```

```python
smallest_pbr_series = filtered_df.groupby('year')['pbr'].nsmallest(15)
smallest_pbr_series
```

```python
selected_index = smallest_pbr_series.index.get_level_values(1)
selected_df = filtered_df.loc[selected_index]
selected_df
```

```python
selected_df.pivot(index='year', columns='Name', values='pbr')
selected_df
```

```python
asset_on_df = selected_df.notna().astype(int).replace(0, np.nan)
```

```python
rtn_series = (yearly_rtn_df * asset_on_df).mean(axis=1)
cum_rtn_str_df = (rtn_series + 1).cumprod()
```

```python
cum_rtn_str_df
```

```python
cum_rtn_str_df.plot()
```
